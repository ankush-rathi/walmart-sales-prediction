```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

train = pd.read_csv("train.csv")

train
```

```python
train['Date'] = pd.to_datetime(train['Date'])
train['Month'] =train['Date'].dt.month 

train['Year'] =train['Date'].dt.year

train['Day'] =train['Date'].dt.day 
train

train['IsHoliday'][train['IsHoliday']==True] = 1
train['IsHoliday'] = train['IsHoliday'].astype(int)

train

feature = pd.read_csv("features.csv")

feature
```

```python
feature['Date'] = pd.to_datetime(feature['Date'])
feature['Month'] =feature['Date'].dt.month 

feature['Year'] =feature['Date'].dt.year

feature['Day'] =feature['Date'].dt.day 
feature

feature['IsHoliday'][feature['IsHoliday']==True] = 1
feature['IsHoliday'] = feature['IsHoliday'].astype(int)

feature
```

```python

feature.isnull().sum()
```

```python
feature['MarkDown1']=feature['MarkDown1'].fillna(feature['MarkDown1'].mean())
feature['MarkDown2']=feature['MarkDown2'].fillna(feature['MarkDown2'].mean())
feature['MarkDown3']=feature['MarkDown3'].fillna(feature['MarkDown3'].mean())
feature['MarkDown4']=feature['MarkDown4'].fillna(feature['MarkDown4'].mean())
feature['MarkDown5']=feature['MarkDown5'].fillna(feature['MarkDown5'].mean())
feature['CPI']=feature['CPI'].fillna(feature['CPI'].mean())
feature['Unemployment']=feature['Unemployment'].fillna(feature['Unemployment'].mean())

feature.isnull().sum()
```

```python

data = pd.merge_asof(train, feature, on='Store',by='Date')
```

```python
print(data.columns.tolist())

data['IsHoliday_y']
data.isnull().sum()

data = data.drop(['IsHoliday_y','Month_y','Year_y','Day_y'],axis=1)

data = data.rename(columns={'IsHoliday_x': 'IsHoliday'})

data = data.rename(columns={'Month_x': 'Month','Year_x':'Year'})
data = data.rename(columns={'Day_x': 'Day'})

data
```

```python
y= data.Weekly_Sales
x=data.drop(['Weekly_Sales','Date'],axis=1)
```

```python

train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2)

model = RandomForestRegressor(n_estimators=100 )
```

```python
model.fit(train_x, train_y.astype(int))
```

```python

y_pred_rf=model.predict(train_x)
acc_rf= round(model.score(train_x, train_y) * 100,3)
print(acc_rf)
```

```python
plt.scatter(train_y, y_pred_rf, alpha=0.5,color='g')
plt.ylabel('Predicted Sale')
plt.xlabel('Actual Sale')
plt.title('Sale Analysis')
plt.xlim([0,220000])
plt.ylim([0,220000])
plt.show()
```

```python
y_pred_rf1=model.predict(test_x)
acc_rf1= round(model.score(test_x, test_y) * 100,3)
print(acc_rf1)
```

```python
plt.scatter(test_y, y_pred_rf1, alpha=0.5,color='g')
plt.ylabel('Predicted Sale')
plt.xlabel('Actual Sale')
plt.title('Sale Analysis')
plt.xlim([0,220000])
plt.ylim([0,220000])
plt.show()
```
